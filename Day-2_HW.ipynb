```python
# Create a menu using while loop

ch = 1

while ch != 0:
    print("\n===== Number Conversion Menu =====")
    ch = int(input("1. Decimal to Binary\n"
                   "2. Decimal to Octal\n"
                   "3. Decimal to Hexadecimal\n"
                   "0. Exit\n\n"
                   "Enter Your Choice: "))

    if ch == 1:
        print("\nDecimal to Binary")
        num = int(input("Enter Decimal Number: "))
        print("Binary =", bin(num)[2:])

    elif ch == 2:
        print("\nDecimal to Octal")
        num = int(input("Enter Decimal Number: "))
        print("Octal =", oct(num)[2:])

    elif ch == 3:
        print("\nDecimal to Hexadecimal")
        num = int(input("Enter Decimal Number: "))
        print("Hexadecimal =", hex(num)[2:].upper())

    elif ch == 0:
        print("\nExit... Thank You!")

    else:
        print("\nInvalid Choice!")
```
