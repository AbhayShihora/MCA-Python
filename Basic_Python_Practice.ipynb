```python
#make a triangle pattern
print("......Program for Drawing Triangle.....\n")
no=int(input("Enter a integer number for draw trianle: "))

for i in range(0,no+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print("\n")
```

```python
#create a menu using while loop


ch=1
while(ch!=0):
    ch=int(input("1.Decimal to Binary \n2.Decimal to octal \n3.Decimal to hexadecimal \n0.Exit \n\nEnter Your choice : "))
    if ch==1:
        print("\tDecimal to Binary")
    elif ch==2:
        print("\tDecimal to octal ")
    elif ch==3:
        print("\tDecimal to hexadecimal")
    elif ch==0:
        print("\tExit...")
    else:
        print("\tInvalid choice!")
```

```python
#Break and Continue statement
print("---Display 0 to 9---")
for i in range(0,10):
    if i==3:
        print("Continue")
        continue
    if i==7:
        print("Break")
        break
    print(i)
```

```python
#make a triangle pattern of alphabat
def alpha_tri(no=5):
    print("......Program for Drawing Triangle.....\n")
    

    for i in range(0,no+1):
        k=65
        for j in range(1,i+1):
            print(chr(k),end=" ")
            k=k+1
        print("\n")
no=int(input("Enter a integer number for draw trianle: "))
alpha_tri(no)#Function call
```

```python
#Fibonaci series
def fibo(no):
    a,b=0,1
    #print(a,b,end=" ")
    while a<no:
        print(a,end=" ")
        a,b=b,a+b
        
    
no=int(input("Enter a number : "))
fibo(no)
```
